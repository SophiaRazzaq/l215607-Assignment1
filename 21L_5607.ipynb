Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1. Each coloumn must contain numbers from between 1 and N, with no duplicates.
2. Each row must contain numbers from between 1 and N, with no duplicates.

```python
# define procedure here

def function(a, n, m):
    if n != m:
      return False
    isPresent = [[False for i in range(n)] 
                        for j in range(m)]
  
    for i in range(n):
        for j in range(m):

            if not isinstance(a[i][j], int):
              return False
            for k in range(n):
                if i != k and a[i][j] == a[k][j]:
                    isPresent[i][j] = True
                    isPresent[k][j] = True
            for k in range(m):
                if j != k and a[i][j] == a[i][k]:
                    isPresent[i][j] = True
                    isPresent[i][k] = True
    for i in range(n):
        for j in range(m):
                if not isPresent[i][j]:
                  continue
                else:
                  return False
    return True
#driver code
#test any value
#n=num of rows
#m=num of cols
#a=the array
n=4
m=4

a = [[1, 2, 3, 4],
        [2, 3, 1, 3],
        [3, 1, 2, 3],
        [4, 4, 4, 4]]
#checker function
flag = function(a, n, m)
print(flag)
```
